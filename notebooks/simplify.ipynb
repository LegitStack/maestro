```python
basics = pd.DataFrame(columns=['input_index','input_value','action_index','action_value','result_index','result_value'])
for index, row in df.iterrows():
    for icol in range(1, 21):
        for acol in range(0, 1):
            temp = df[(df[('input', icol)] == row[('input', icol)])&(df[('action', acol)] == row[('action', acol)])]
            if temp.shape[0] > 1:
                print(temp)
                for rcol in range(1, 21):
                    answer = temp[('result', rcol)].unique()
                    if answer.shape[0] == 1:
                        basics = basics.append({'input_index':icol, 'input_value':row[('input', icol)], 'action_index':acol, 'action_value':row[('action', 0)], 'result_index':rcol, 'result_value':row[('result', rcol)]},ignore_index=True)
                        #basics[(icol, row[('input', icol)], acol, row[('action', 0)])] = (rcol, row[('result', rcol)])
    
```
